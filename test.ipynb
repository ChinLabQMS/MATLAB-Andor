```matlab
data_path = 'data/r=50_darkspot_onatoms_centered.mat';
Data = load(data_path).Data;
Data
```

```matlab
Data.SequenceTable
```

```matlab
images = Data.Andor19330.Image;
mean_image = mean(images, 3);

figure
imagesc(mean_image);
daspect([1 1 1]);
title('Mean Image')
colorbar;
```

```matlab
x_sum = sum(mean_image, 1);
y_sum = sum(mean_image, 2);

figure
subplot(2, 1, 1)
plot(x_sum)
title('X Sum')
subplot(2, 1, 2)
plot(y_sum)
title('Y Sum')
```
